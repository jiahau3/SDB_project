```python
import rioxarray as rxr
import pyproj
from pystac_client import Client
import stackstac
import xarray as xr
import numpy as np

%matplotlib inline
```

#### Load bathymetric data from bathybase dataset

```python
swanson_tiff = rxr.open_rasterio('../../preprocessing/data/bathybase/bathy_tiffs/swanson_reservoir.tiff')
```

```python
swanson_tiff
```

```python
wgs84 = pyproj.CRS('EPSG:4326')
```

#### Get the bounding box coordinate and reproject into WGS84

```python
bounding_box = swanson_tiff.rio.transform_bounds(wgs84)
left, bottom, right, top = bounding_box
latitude = (bottom, top)
longitude = (left, right)
```

```python
bounding_box
```

#### Create a Client instance for the STAC API

```python
aws_stac_url = "https://earth-search.aws.element84.com/v0"
catalog = Client.open(aws_stac_url)
```

#### Key in the criteria for searching satellite images. The 1 yr data with cloud cover rate less than 10% are collected

```python
%%time
collections = ["sentinel-s2-l2a-cogs"]
items = catalog.search(collections=collections, bbox=bounding_box, datetime="2020-10-01/2021-10-01",query={"eo:cloud_cover": {"lt": 10}}).get_all_items()
len(items)
```

#### Turn the collection into DataArray

```python
stack = stackstac.stack(items)
```

```python
stack
```

#### After images are collected, we slice the region we want to observe by getting the bounding box with the same CRS as images 

```python
minx, miny, maxx, maxy = swanson_tiff.rio.transform_bounds(stack.crs)
minx, miny, maxx, maxy
```

```python
aoi = stack.sel(x=slice(minx, maxx), y=slice(maxy,miny))
aoi
```

#### Check visual bands by plotting

```python
visual=aoi.sel(band=["B04", "B03", "B02"], time='2020-10-07')
```

```python
visual.isel(time=0).plot.imshow(rgb="band", robust=True, size=6)
```

#### Eliminate the cloud by taking median value in the 1 yr collection data

```python
aoi_median = aoi.median(dim="time")
aoi_median
```

```python
aoi_median.sel(band=['B04', 'B03', 'B02']).plot.imshow(rgb="band", robust=True, size=6)
```

#### Exclude overview, true color channels both of which are not single band

```python
%%time
aoi_median = aoi_median.sel(band=['B01', 'B02', 'B03', 'B04', 'B05', 'B06', 'B07',
                     'B08', 'B8A', 'B09', 'B11', 'B12', 'AOT', 'WVP', 'SCL']).compute()
```

```python
aoi_median.rio.resolution()
```

```python
bands, height, width = aoi_median.shape
```

```python
aoi_median.isel(band=1).plot()
```

```python
stack.rio.crs
```

#### Downsample the depth data from 5m to 10m and reprojecting to the same CRS as satellite data

```python
swanson_tiff_10m = swanson_tiff.rio.reproject(stack.rio.crs, shape=(height, width), resampling=1)
```

```python
swanson_tiff_10m
```

```python
swanson_tiff_10m.plot()
```

#### Conform the depth data onto median data

```python
swanson_depth = swanson_tiff_10m.sel({'band': 1}).reindex_like(aoi_median.sel(band='B04'), method='nearest')
swanson_depth.name = aoi_median.name
swanson_depth = swanson_depth.assign_coords({"band": "bathy"})
```

#### Combine all the remaining bands with depth data

```python
combined_swanson = xr.concat([aoi_median.reset_coords().isel(band=np.arange(aoi_median.shape[0]))[aoi_median.name], 
                              swanson_depth], dim="band")
```

```python
combined_swanson
```

#### Clip off the data which is not in the depth region

```python
masked_combined = combined_swanson.where(combined_swanson.sel({'band': "bathy"}).notnull())
masked_combined
```

```python
masked_combined.sel({'band': 'B04'}).plot()
```

#### Save the processed data

```python
from pathlib import Path
dirpath = "../data/swanson/"
Path(dirpath).mkdir(parents=True, exist_ok=True)
```

```python
filename = '202010_1yr_median_clipped.tif'
masked_combined.rio.to_raster(dirpath+filename)
```
